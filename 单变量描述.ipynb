```python
#单变量描述统计

##无序类别变量描述统计
import pandas as pd  
import matplotlib.pyplot as plt  
plt.rcParams['font.sans-serif']=['SimHei']#设置字体  
df = pd.read_spss("identity.sav")
```

```python
df['民族'].value_counts(normalize=True)
```

```python
x=df['政治面貌'].value_counts().index
y=df['政治面貌'].value_counts(normalize=True).values * 100
fig, ax2 = plt.subplots()  
labels = df['政治面貌'].value_counts().index  
ax2.pie(y, autopct='%1f%%', labels=labels, textprops=dict(color='w'))  
plt.show()
```
